```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get("https://bookpage.com/features")
page
```

```python
page.content
```

```python
soup=BeautifulSoup(page.content,'html.parser')
soup
```

```python
Name=soup.find_all('h4', class_="italic")
Name
```

```python
Book_Names=[]
for i in Name:
    Book_Names.append(i.text)
Book_Names
```

```python
Book_Names=[]
for i in Name:
    Book_Names.append(i.text.replace('\n',''))
Book_Names
```

```python
Author=soup.find_all('p', class_="sans bold")
Author
```

```python
Book_Authors=[]
for i in Author:
    Book_Authors.append(i.text)
Book_Authors
```

```python
Book_Authors=[]
for i in Author:
    Book_Authors.append(i.text.replace('\n',''))
Book_Authors
```

```python
Genre=soup.find_all('p', class_="genre-links hidden-phone")
Genre
```

```python
Book_Genres=[]
for i in Genre:
    Book_Genres.append(i.text)
Book_Genres
```

```python
Book_Genres=[]
for i in Genre:
    Book_Genres.append(i.text.replace('\n',''))
Book_Genres
```

```python
Review=soup.find_all('p', "excerpt")
Review
```

```python
Book_Reviews=[]
for i in Review:
    Book_Reviews.append(i.text)
Book_Reviews
```

```python
Book_Reviews=[]
for i in Review:
    Book_Reviews.append(i.text.replace('\n',''))
Book_Reviews
```

```python
import pandas as pd
BookPage=pd.DataFrame({})
BookPage['Book_Names']=Book_Names[:5]
BookPage['Book_Authors']=Book_Authors[:5]
BookPage['Book_Reviews']=Book_Reviews[:5]
BookPage['Book_Genres']=Book_Genres[:5]
BookPage
```

```python
import pickle
filename= 'BookPage.pkl'
pickle.dump(BookPage,open(filename,'wb'))
```
